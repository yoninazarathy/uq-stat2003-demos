# Some key continuous probability distributions with SciPy

See: https://docs.scipy.org/doc/scipy/reference/stats.html

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma, beta, norm, pareto, weibull_min, cauchy

def do_plot():
    plt.plot(x, pdf1, label = label1); 
    plt.plot(x, pdf2, label = label2); 
    plt.title(dist_name); 
    plt.xlabel('x'); 
    plt.ylabel('f(x)'); 
    plt.grid(); 
    plt.legend()
    plt.show()
```

# Normal Distribution

See: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html#scipy.stats.norm

```python
mu1, sigma1 = 0, 1
mu2, sigma2 = 0.25, 1.5
label1 = f"mu = {mu1}, sigma = {sigma1}"
label2 = f"mu = {mu2}, sigma = {sigma2}"
dist_name = f"Normal"
x = np.linspace(-6, 6, 1000)
pdf1 = norm.pdf(x, mu1, sigma1)
pdf2 = norm.pdf(x, mu2, sigma2)
do_plot()
```

# Gamma Distribution
See: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.gamma.html

```python
shape1, scale1 = 2.3, 1.0
shape2, scale2 = 1.8, 1.0
label1 = f"shape = {shape1}, scale = {scale1}"
label2 = f"shape = {shape2}, scale = {scale2}"
dist_name = f"Gamma"
x = np.linspace(0, 15, 1000)
pdf1 = gamma.pdf(x, shape1, scale=scale1)
pdf2 = gamma.pdf(x, shape2, scale=scale2)
do_plot()
```

# Beta Distribution

See: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.beta.html#scipy.stats.beta

```python
a1, b1 = 0.99, 1
a2, b2 = 1, 0.9
label1 = f"a = {a1}, b = {b1}"
label2 = f"a = {a2}, b = {b2}"
dist_name = f"Beta"
x = np.linspace(0, 1, 1000)
pdf1 = beta.pdf(x, a1, b1)
pdf2 = beta.pdf(x, a2, b2)
do_plot()
```

# Pareto Distribution

See: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.pareto.html#scipy.stats.pareto

```python
b1 = 3
b2 = 5
label1 = f"b = {b1}"
label2 = f"b = {b2}"
dist_name = f"Pareto"
x = np.linspace(0, 5, 1000)
pdf1 = pareto.pdf(x, b1)
pdf2 = pareto.pdf(x, b2)
do_plot()
```

# Cauchy Distribution (vs. Standard Normal)

See: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.cauchy.html#scipy.stats.cauchy

```python
label1 = f"Standard Normal"
label2 = f"Cauchy"
dist_name = f"Cauchy vs. Normal"
x = np.linspace(-6, 6, 1000)
pdf1 = norm.pdf(x, 0, 1)
pdf2 = cauchy.pdf(x)
do_plot()
```
